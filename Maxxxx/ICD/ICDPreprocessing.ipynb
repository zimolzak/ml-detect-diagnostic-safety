```python
import pandas as pd
```

```python
csv_names = ["Refined_SPADE_RiskFactors_dizziness_ICD9codes.csv", "Refined_SPADE_RiskFactors_dizziness_ICD10codes.csv", 
            "Refined_SPADE_RiskFactors_headache_ICD9codes.csv", "Refined_SPADE_RiskFactors_headache_ICD10codes.csv",
            "Refined_SPADE_RiskFactors_stroke_ICD9codes.csv", "Refined_SPADE_RiskFactors_stroke_ICD10codes.csv",
            "Refined_SPADE_RiskFactors_riskfactor_ICD9codes.csv", "Refined_SPADE_RiskFactors_riskfactor_ICD10codes.csv"]
```

```python
def convert_csv(name):
    df = pd.read_csv(name)
    icd_col = df.columns[0]
    df[icd_col] = df[icd_col].astype(str)
    df[icd_col] = df[icd_col].str[0:3] + "." + df[icd_col].str[3:]
    return df
```

```python
for name in csv_names:
    df = convert_csv(name)
    df.to_csv("p_" + name)
```
