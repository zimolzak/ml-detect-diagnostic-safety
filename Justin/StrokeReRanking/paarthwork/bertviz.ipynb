```python
from transformers import AutoTokenizer, AutoModel
from bertviz import head_view
```

```python
tokenizer = AutoTokenizer.from_pretrained("C:/Users/vhahoumowerj/.cache/huggingface/transformers/Bio_ClinicalBERT/")
model=AutoModel.from_pretrained('C:/Users/vhahoumowerj/.cache/huggingface/transformers/Bio_ClinicalBERT/', output_attentions=True)
inputs = tokenizer.encode("The patient presented with acute MI.", return_tensors='pt')
outputs = model(inputs)
attention = outputs[-1]
tokens = tokenizer.convert_ids_to_tokens(inputs[0])
```

```python
head_view(attention, tokens)
```
